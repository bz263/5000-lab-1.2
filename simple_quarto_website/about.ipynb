```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x) + 0.25*np.cos(3*x)

plt.figure()
plt.plot(x, y)
plt.title("Demo Plot: sin(x) + 0.25 cos(3x)")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```
